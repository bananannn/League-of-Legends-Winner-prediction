##  Champions Info Aggregation

```python
champ_win_rate = merged.groupby(['championid', 'name']).agg({'win': 'sum', 'name': 'count', 'kills': 'mean', 'deaths': 'mean', 'assists': 'mean'})
champ_win_rate.columns = ['win_matches', 'total_games', 'kills', 'deaths', 'assists']

champ_win_rate['win_rate'] = champ_win_rate['win_matches'] / champ_win_rate['total_games'] * 100
champ_win_rate['pick_rate'] = champ_win_rate['total_games'] / champ_win_rate['total_games'].sum() * 100
champ_win_rate['champion_score'] = (champ_win_rate['kills'] + champ_win_rate['assists'])/champ_win_rate['deaths']
champ_win_rate = champ_win_rate.sort_values('championid', ascending = True)

champ_win_rate.reset_index(inplace=True)
    
```

```python
champ_win_rate = champ_win_rate.sort_values('win_rate', ascending = True)

champ_win_rate
```

```python
from bokeh.plotting import figure, output_file, show

#import bokeh.io as io
from bokeh.models import ColumnDataSource, Range1d, LabelSet, Label

#io.output_notebook()
output_file("label.html", title="label.py example")

source = ColumnDataSource(data=dict(win_rate = list(champ_win_rate['win_rate']),
                                    champion_score = list(champ_win_rate['champion_score']),
                                    names = list(champ_win_rate['name'])))

p = figure(plot_width=1200, plot_height=800)

# p.toolbar.logo = "grey"
#p.background_fill_color = "#dddddd"
#p.scatter(x='win_rate', y='champion_score', size=8, source=source)
p.xaxis[0].axis_label = "win rate (%)"
p.yaxis[0].axis_label = "champion score"
# p.grid.grid_line_color = "white"

p.circle(x = 'win_rate', y = 'champion_score', size=12, source = source,color = '#f4a582', line_color="black", fill_alpha=0.6)

labels = LabelSet(x='win_rate', y='champion_score', text='names',
              x_offset=5, y_offset=5, source=source, text_font_size="8pt",render_mode='canvas')

p.add_layout(labels)

show(p)
```
