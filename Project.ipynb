# Data Preprocessing

```python
import pandas as pd

matches = pd.read_csv('data/matches.csv')
champ = pd.read_csv('data/champs.csv')
parti = pd.read_csv('data/participants.csv')
```

```python
stats1 = pd.read_csv('data/stats1.csv')

stats2 = pd.read_csv('data/stats2.csv')

stats = stats1.append(stats2)

stats.reset_index(inplace=True)

bans = pd.read_csv('data/teambans.csv')
teamstats = pd.read_csv('data/teamstats.csv')

stats.head(10)
```

```python
merged = pd.merge(parti, stats, how = 'left', on = ['id'], suffixes=('', '_y'))
merged = pd.merge(merged, champ, how = 'left', left_on = 'championid', right_on = 'id', suffixes=('', '_y'))
merged = pd.merge(merged, matches, how = 'left', left_on = 'matchid', right_on = 'id', suffixes=('', '_y'))
```

```python
def combine_position(row):
    if row['role'] in ('DUO_SUPPORT', 'DUO_CARRY'):
        return row['role']
    else:
        return row['position']
    
    
merged['final_position'] = merged.apply(combine_position, axis = 1) 
```

##  Champions Info Aggregation

```python
champ_win_rate = merged.groupby(['championid', 'name']).agg({'win': 'sum', 'name': 'count', 'kills': 'mean', 'deaths': 'mean', 'assists': 'mean'})
champ_win_rate.columns = ['win_matches', 'total_games', 'kills', 'deaths', 'assists']

champ_win_rate['win_rate'] = champ_win_rate['win_matches'] / champ_win_rate['total_games'] * 100
champ_win_rate['pick_rate'] = champ_win_rate['total_games'] / champ_win_rate['total_games'].sum() * 100
champ_win_rate['champion_score'] = (champ_win_rate['kills'] + champ_win_rate['assists'])/champ_win_rate['deaths']
champ_win_rate = champ_win_rate.sort_values('championid', ascending = True)

champ_win_rate.reset_index(inplace=True)
    
```

```python
champ_win_rate = champ_win_rate.sort_values('win_rate', ascending = True)

champ_win_rate.head(10)
```

```python
def creat_X_Y_from_topk(datalist, k):
    X = []
    Y = []
    for i in range(k):
        nan_exist = False
        for j in datalist[i]:
            if math.isnan(j):
                nan_exist = True
        if nan_exist == True:
            continue
        X.append(datalist[i][1:])
        Y.append(int(datalist[i][0]))
    return np.asarray(X),np.asarray(Y) 

[X,Y] = creat_X_Y_from_topk(datalist, 200000)    
```

```python
from sklearn.preprocessing import StandardScaler
def generate_data_std(X,Y,train_ratio=0.7):
    
    split_index = int(train_ratio * len(X))
    X_train = X[0:split_index]
    Y_train = Y[0:split_index]
    X_test = X[split_index:]
    Y_test = Y[split_index:]
    
    sc=StandardScaler()
    X_train_std=sc.fit_transform(X_train)
    X_test_std=sc.transform(X_test)
    
    print(X_train_std.shape)
    return X_train_std, Y_train, X_test_std, Y_test
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
import math
import requests
import itertools
import time
import sys
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.externals import joblib
from sklearn import tree
from sklearn.naive_bayes import MultinomialNB
import pylab

def GBC_model():
    start = time.time()
    print("GBC_model function start:", start)
    
    params = {'n_estimators':range(10, 300, 50), 'learning_rate':np.linspace(0.001, 0.1, 5),
        'max_depth':range(1, 30, 4), 'min_samples_leaf':range(1, 30, 4)}
#     params = {'n_estimators':range(10, 300, 3), 'learning_rate':np.linspace(0.001, 0.1, 3),
#         'max_depth':range(1, 30, 3), 'min_samples_leaf':range(1, 30, 3)}
    model = GradientBoostingClassifier(min_samples_split=15,max_features='sqrt',
                                       subsample=0.8,random_state=189)
    
    end = time.time()
    print("GBC_model function end:", end - start)
    return model, params

def MLPC_model():
    start = time.time()
    print("MLPC_model function start:", start)
    model = MLPClassifier(random_state=189, max_iter=1000)
    params = {'hidden_layer_sizes':[(100), (100, 100), (50, 100), (100, 50), (100, 100, 100),
                                    (50, 100, 50), (100, 50, 100), (100, 100, 50), (50, 50, 100),
                                    (100, 100, 100, 100), (100, 50, 100, 50), (50, 100, 50, 100),
                                    (100, 100, 100, 100, 100), (100, 50, 100, 50, 100),
                                    (50, 100, 50, 100, 50), (100, 100, 50, 100, 100),
                                    (100, 100, 100, 100, 100, 100), (100, 100, 100, 100, 100, 100, 100),
                                    (80, 70, 60, 50, 40, 30, 20, 10), (200, 150, 100, 50, 20),
                                    (120, 90, 75, 63, 55, 50)],
              'activation':['relu', 'tanh'], 'alpha':np.linspace(0.000001, 0.001, 10)}
    
    end = time.time()
    print("MLPC_model function end:", end - start)
    return model, params

def RFC_model():
    start = time.time()
    print("RFC_model function start:", start)
    params = {'criterion':['gini', 'entropy'], 'n_estimators':range(10,311,100),
              'max_features':['sqrt','log2',0.2,0.4,0.6,0.8], 'max_depth':range(3,25,10),
              'min_samples_split':range(5,30,10), 'min_samples_leaf':range(5,30,10)}
    model = RandomForestClassifier(random_state=189)
    
    end = time.time()
    print("RFC_model function end:", end - start)
    return model, params

def SVC_model():
    params = {'C':np.linspace(0.01,5,25), 'kernel':['linear','poly','rbf','sigmoid']}
    model = SVC(max_iter=5000,random_state=189)
    return model, params

def LR_model():
    params = {'C':np.linspace(0.01,5,10),
            'solver':['newton-cg','lbfgs','liblinear','sag','saga']}
    model = LogisticRegression(max_iter=1000, random_state=189)
    return model, params

def DTC_model():
    model=tree.DecisionTreeClassifier()
    return model


def grid_search_init(model_name):
    
    
    print("Model: ", model_name)
    X_train, Y_train, X_test, Y_test = generate_data_std(X,Y)
    if model_name == "GBC":
        model, params = GBC_model()
    if model_name == "MLPC":
        model, params = MLPC_model()
    if model_name == "RFC":
        model, params = RFC_model()
    if model_name == "SVC":
        model, params = SVC_model()
    if model_name == "LRC":
        model, params = LR_model()
    if model_name=='DTC':
        model=DTC_model()


    model.fit(X_train,Y_train)
    if model_name == "LRC":
        print(model.coef_)

    y_pred = [i for i in model.predict(X_test)]
    val_accuracy = sum([1 if i == j else 0 for i, j in zip(y_pred, Y_test)])/len(y_pred)
    print("Validation Accuracy:", val_accuracy)
    return val_accuracy

models = ['RandomForestClassifier', 'GradientBoostingClassifier', 'MLPClassifier', 'SVC', 'LogisticRegression', 'DecisionTree']
validation_accuracy = []
rfc = grid_search_init("RFC")
validation_accuracy.append(rfc)

gbc = grid_search_init("GBC")
validation_accuracy.append(gbc)

mlpc = grid_search_init("MLPC")
validation_accuracy.append(mlpc)

svc = grid_search_init("SVC")
validation_accuracy.append(svc)

lrc = grid_search_init("LRC")
validation_accuracy.append(lrc)

dtc = grid_search_init("DTC")
validation_accuracy.append(dtc)
```
